# Analysis of the transformer experiments

```python
import os
import shutil
import respiration.utils as utils

models_dir = utils.dir_path('models', 'transformer')
logs_dir = utils.dir_path('outputs', 'logs')

# List all the files in the model directory
models = os.listdir(models_dir)

metadata = []

# Check if the model directory contains a manifest.json file
for model in models:
    model_dir = utils.join_paths(models_dir, model)
    log_dir = utils.join_paths(logs_dir, model)
    files = os.listdir(model_dir)

    # Delete the model directory if it is empty
    if 'manifest.json' not in files:
        print(f'Deleting {model}')
        shutil.rmtree(model_dir)
        shutil.rmtree(log_dir, ignore_errors=True)

    # Read the manifest file
    manifest_path = utils.join_paths(model_dir, 'manifest.json')
    manifest = utils.read_json(manifest_path)
    metadata.append(manifest)
```

```python
import pandas as pd

df = pd.DataFrame(metadata)

# Display all rows where the loss_fn is MSELoss
df[df['loss_fn'] == 'MSELoss']
```
